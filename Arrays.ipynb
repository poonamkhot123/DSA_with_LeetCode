```python
def moveZeros(nums):
  if len(nums) == 1:
    return nums
  i = 0
  while i < len(nums):
    if nums[i] == 0:
      break
    i+=1
  if i ==len(nums):
    return nums
  j=i+1
  while j < len(nums):
    if nums[j]!=0:
      nums[i],nums[j]=nums[j],nums[i]
      i+=1
    j+=1
  return nums
# Example:
nums = [0, 1, 0, 3, 12]
print(moveZeros(nums))  # Output: [1, 3, 12, 0, 0]
```

```python
def majorityElement(nums):
    count = 0
    candidate = None

    for num in nums:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)

    return candidate

s = [2,2,1,1,1,2,2]
print(majorityElement(s))
```

```python
# remove duplicates from sorted array
def removeDuplicates(nums):
  n = len(nums)
  start = 0
  for i in range(1,n):
    # unique element
    if nums[i] != nums[start]:
      start +=1
      nums[start]=nums[i]
  return nums[:start+1]  # return only unique element.

s = [1,2,3,4,5,5,6,6,6,7]
print(removeDuplicates(s))
```

```python
# Best time to buy and sell stock - 1
def maxProfit(prices):
  min_price = prices[0]
  profit = 0
  for i in range(1,len(prices)):
    curr_profit = prices[i] - min_price

    if curr_profit > profit:
      profit = curr_profit
    min_price = min(min_price,prices[i])

  return profit
s = [7,1,5,3,6,4]
print(maxProfit(s))
```

```python
# Best time to buy and sell stock - II
def maxProfit(prices):
  profit = 0
  n = len(prices)
  for i in range(n-1):
    if prices[i+1] >prices[i]:
      profit += (prices[i+1]-prices[i])
  return profit

s = [7,1,5,3,6,4]
print(maxProfit(s))
```

```python
def rotateArray(nums,k):
  n = len(nums)
  rotations = k%n
  for i in range(0,rotations):
      e = nums.pop()
      nums.insert(0,e)
  return nums
nums = [1,2,3,4,5,6,7]

print(rotateArray(nums,3))
```

```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1] * n

    # Step 1: Calculate prefix product
    prefix = 1
    for i in range(n):
        answer[i] = prefix
        prefix *= nums[i]

    # Step 2: Calculate suffix product (from right)
    suffix = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= nums[i]

    return answer

# Example
nums = [1, 2, 3, 4]
print(productExceptSelf(nums))
```

```python
class Solution:
    def increasingTriplet(self, nums):
        first = float('inf')
        second = float('inf')

        for n in nums:
            if n <= first:
                first = n            # smallest number
            elif n <= second:
                second = n           # second smallest number
            else:
                # found a number greater than both → triplet exists
                return True

        return False
nums1 = [1, 2, 3, 4, 5]
nums2 = [5, 4, 3, 2, 1]
nums3 = [2, 1, 5, 0, 4, 6]

sol = Solution()
print(sol.increasingTriplet(nums1))  # True
print(sol.increasingTriplet(nums2))  # False
print(sol.increasingTriplet(nums3))  # True
```

```python
def firstMissingPositive(nums):
    n = len(nums)

    for i in range(n):
        # Place nums[i] in its correct position
        while 1 <= nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            # Swap nums[i] with nums[nums[i]-1]
            nums[nums[i] - 1], nums[i] = nums[i], nums[nums[i] - 1]

    # Check for first missing positive
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1

    # If all are in place, smallest missing is n+1
    return n + 1
nums = [3, 4, -1, 1]
print(firstMissingPositive(nums))
```
