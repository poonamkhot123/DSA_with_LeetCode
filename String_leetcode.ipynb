```python
# by using two pointer.
def isSubsequence(s, t):
    i = 0  # pointer for s
    j = 0  # pointer for t

    while i < len(s) and j < len(t):
        if s[i] == t[j]:
            i += 1
        j += 1

    return i == len(s)

s = "abc"
t = "ahbgdc"

print(isSubsequence(s, t))
```

```python
def isPalindrome(s):
    left, right = 0, len(s) - 1

    while left < right:
        # skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1
        # skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1

        # compare lowercase characters
        if s[left].lower() != s[right].lower():
            return False

        left += 1
        right -= 1

    return True
s = "A man, a plan, a canal: Panama"
print(isPalindrome(s))
```

```python
def longestCommonPrefix(strs):
    if not strs:
        return ""

    prefix = strs[0]

    for s in strs[1:]:
        while not s.startswith(prefix):
            prefix = prefix[:-1]
            if prefix == "":
                return ""

    return prefix

# Example 1
strs1 = ["flower", "flow", "flight"]
print("Input:", strs1)
print("Longest Common Prefix:", longestCommonPrefix(strs1))
print()

# Example 2
strs2 = ["dog", "racecar", "car"]
print("Input:", strs2)
print("Longest Common Prefix:", longestCommonPrefix(strs2))
print()

# Example 3
strs3 = ["interview", "internet", "internal", "interval"]
print("Input:", strs3)
print("Longest Common Prefix:", longestCommonPrefix(strs3))
```

```python
def convert(s, numRows):
    if numRows == 1 or numRows >= len(s):
        return s

    rows = [''] * numRows
    curr_row = 0
    going_down = False

    for char in s:
        rows[curr_row] += char
        # Change direction at top or bottom
        if curr_row == 0 or curr_row == numRows - 1:
            going_down = not going_down
        curr_row += 1 if going_down else -1

    return ''.join(rows)

# Example 1
s1 = "PAYPALISHIRING"
numRows1 = 3
print("Input:", s1, "NumRows =", numRows1)
print("Output:", convert(s1, numRows1))
print()

# Example 2
s2 = "PAYPALISHIRING"
numRows2 = 4
print("Input:", s2, "NumRows =", numRows2)
print("Output:", convert(s2, numRows2))
```

```python
def reverseWords(s):
    # Split by spaces, remove extra spaces
    words = s.split()
    # Reverse the list of words
    reversed_words = words[::-1]
    # Join with single space
    return ' '.join(reversed_words)

# Example 1
s1 = "the sky is blue"
print("Input:", s1)
print("Output:", reverseWords(s1))
print()

```

```python
# Simulated master class for testing
class Master:
    def __init__(self, secret):
        self.secret = secret
        self.guesses = 0

    def guess(self, word):
        self.guesses += 1
        return sum(c1 == c2 for c1, c2 in zip(word, self.secret))

# Easy guess function
def findSecretWord(wordlist, master):
    for _ in range(10):
        # Pick the first word from the list
        guess_word = wordlist[0]
        matches = master.guess(guess_word)
        print(f"Guess: {guess_word}, Matches: {matches}")
        if matches == len(master.secret):
            print("Secret word found!")
            return
        # Keep only words that have same number of matches
        wordlist = [w for w in wordlist if sum(c1 == c2 for c1, c2 in zip(w, guess_word)) == matches]

# Example
wordlist = ["acckzz","ccbazz","eiowzz","abcczz"]
secret = "acckzz"
master = Master(secret)

findSecretWord(wordlist, master)
```
