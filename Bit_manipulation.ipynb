```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        count = 0
        while n:
            count += n & 1     # Check if last bit is 1
            n >>= 1            # Shift bits to the right
        return count
print(Solution().hammingWeight(11))
```

```python
class Solution:
    def countBits(self, n: int):
        ans = [0] * (n + 1)
        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)
        return ans
print(Solution().countBits(5))
```

```python
class Solution:
    def reverseBits(self, n: int) -> int:
        result = 0
        for i in range(32):             # There are 32 bits
            bit = n & 1                 # Get the last bit of n
            result = (result << 1) | bit  # Add it to the result
            n >>= 1                     # Shift n to the right by 1
        return result
print(Solution().reverseBits(43261596))
```

```python
# LeetCode 201: Bitwise AND of Numbers Range

# Define the class and function
class Solution:
    def rangeBitwiseAnd(self, left: int, right: int) -> int:
        shift = 0
        # Keep shifting right until left == right
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1
        # Shift back to original position
        return left << shift


# Example 1
left = 5
right = 7
result = Solution().rangeBitwiseAnd(left, right)
print(f"Input: left = {left}, right = {right}")
print(f"Output: {result}\n")   # Expected: 4

```

```python
# LeetCode 260: Single Number III

class Solution:
    def singleNumber(self, nums):
        # Step 1: XOR all numbers to get xor_all = a ^ b
        xor_all = 0
        for num in nums:
            xor_all ^= num

        # Step 2: Find rightmost set bit (difference bit)
        diff_bit = xor_all & -xor_all

        # Step 3: Divide numbers into two groups and XOR each
        num1 = num2 = 0
        for num in nums:
            if num & diff_bit:
                num1 ^= num
            else:
                num2 ^= num
        return [num1, num2]


#  Example 1
nums = [1, 2, 1, 3, 2, 5]
result = Solution().singleNumber(nums)
print("Input:", nums)
print("Output:", result)
```

```python
class Solution:
    def getSum(self, a: int, b: int) -> int:
        # 32-bit mask for overflow handling
        MASK = 0xFFFFFFFF
        INT_MAX = 0x7FFFFFFF

        while b != 0:
            # Sum without carry & carry calculation
            a, b = (a ^ b) & MASK, ((a & b) << 1) & MASK

        # Convert to Python's signed integer
        return a if a <= INT_MAX else ~(a ^ MASK)


#  Example Runs
obj = Solution()

# Example 1
a, b = 1, 2
print(f"Input: a = {a}, b = {b}")
print("Output:", obj.getSum(a, b))   # Expected: 3

# Example 2
a, b = -2, 3
print(f"\nInput: a = {a}, b = {b}")
print("Output:", obj.getSum(a, b))   # Expected: 1

# Example 3
a, b = -5, -7
print(f"\nInput: a = {a}, b = {b}")
print("Output:", obj.getSum(a, b))   # Expected: -12
```
